### Import salaries data for San Francisco

### Import an online data set
- Sometimes you can import online data sets into Python without the need to download it to your computer as long as you have the link
- Import the data set from this online source "http://672258.youcanlearnit.net/college.csv" and call this object "college"

```python
# Import data
url = 'http://672258.youcanlearnit.net/college.csv'
college = pd.read_csv(url)
# Show the data set's information
college.info
# Show the first 10 observations 
college.head(10)
```

```python
# Find the standard deviation of the admission rate
college['admission_rate'].std()
```

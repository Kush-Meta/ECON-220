## Sampling Distribution of Sample Mean
### With salary data

```python
# Extract variable "salary2019"
salary = salary_data["salary2019"]
#Convert into array
salary=np.array(salary)
```

```python
# Plot distribution of salary
sns.histplot(salary);
```

```python
# Plot distribution of salary
sns.histplot(salary, bins=20);
```

```python
# simulation: think about salary as population ->generate distribution og sample mean of salary
# Number of iterations 
itars = 10000
# Empty vector to store the simulated standard means
sampling_mean=[]
# Loop to generate sample means
for i in range(0,itars):
    random_sample = np.random.choice(salary, 30)
    sample_mean = np.mean(random_sample)
    sampling_mean.append(sample_mean)
```

```python
# Plot the distribution of the sample means
sns.histplot(sampling_mean);
```

## Sampling Distribution of sample proportion

```python
# Unemployment rate: 3.0%
# Find probability that in sample of 100 people, 3 or fewer are unemployed
stats.binom.cdf(3,100,0.030)
```

```python
# Simulation: generate sample proportions
# Number of iterations 
itars = 10000
# Empty vector to store the simulated standard means
unemp=[]
# Loop to generate sample means
for i in range(0,itars):
    random_sample = stats.binom.rvs(n=1000,p=0.038)
    proportion = random_sample/1000
    unemp.append(proportion)
```

```python
# Plot the distribution of the sample proportions
sns.histplot(unemp, bins=10);
```

```python
# Mean and standard deviation of distribution above
mean_unemp = np.mean(unemp)
std_unemp = np.std(unemp)
```

```python
# Probability that 3 or fewer people are unemployed
# -> Probability that less than 3% of people are unemployed
stats.norm.cdf(0.03,loc=mean_unemp, scale=std_unemp)
```

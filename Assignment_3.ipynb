# Probability Exercises

Find the probability that a normal random variable with a mean 10 and a standard deviation of 7
* (i) lies between 9 and 12
* (ii) is greater than 25
* (iii) is less than 13

```python
# Use print() for each command to show your work.
mean = 10
std_dev = 7

# (i) Probability that the random variable lies between 9 and 12
prob_1 = stats.norm.cdf(12, mean, std_dev) - stats.norm.cdf(9, mean, std_dev)

# (ii) Probability that the random variable is greater than 25
prob_2 = 1 - stats.norm.cdf(25, mean, std_dev)

# (iii) Probability that the random variable is less than 13
prob_3 = stats.norm.cdf(13, mean, std_dev)

# Printing the results
print("Probability between 9 and 12:", prob_1.round(3))
print("Probability greater than 25:", prob_2.round(3))
print("Probability less than 13:", prob_3.round(3))
```

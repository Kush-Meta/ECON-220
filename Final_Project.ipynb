## Introduction: 
Our research question for this report aims to compare the trend between the value of the house and the total personal income. This topic is relevant as it addresses a key aspect of income inequality in the economy. However, the house value to income ratio differed across cities, with the highest ratios being on the West Coast, especially Honolulu (12.1), San Jose (12.0), and San Francisco (11.3) (Hermann & Whitney, 2024).

It is important to analyze these trends to understand whether higher incomes truly correlate with higher house values or if other factors are affecting this ratio and the increase in the value of houses in the US.

## Literature Review:
Article 1:
The Effects of Homeownership: on the Self-Esteem, Perceived Control and Life Satisfaction of Low-Income People
By William M. Rohe & Michael A. Stegman

- Homeowners did not have higher self-esteem compared to themselves prior to purchasing a home. However, they had a higher life satisfaction compared to before purchasing a home. The study looked at low-income homeowners, showing the psychological effects of house ownership on them. However, housing conditions affected both self-esteem and life satisfaction. Better housing conditions were found among higher-income people. 

Article 2:
Homeownership, wealth accumulation and income status
By Tracy M. Turner & Heather Luea

- Home ownership increases average wealth by 13,700 dollars for low-income people. In comparison, home ownership increases average wealth by $15,000 for high-income people. This shows how higher-income homeowners derive more benefits from owning a house. However, people of all income levels benefit from owning a home. When the country went through a period of wealth accumulation, both lower-income and higher-income benefited from the wealth creation effects of house ownership. 

Article 3:
The social and economic consequences of housing wealth inequalities
By Adriana Mihaela Soaita, Duncan Maclennan, and Kenneth Gibb

- Housing wealth inequality refers to the disparity in housing worth among a defined group. In deregulated market-based housing systems there is more housing wealth inequality. This occurs in market-based systems as owning a house is seen as an investment. Additionally, housing wealth can benefit middle-income people during a crisis, low-income people do not have this luxury. With rising house prices, younger people can purchase housing, leading to growing housing wealth inequality among different age groups. 

Article 4:
Aging and the income value of housing wealth
By Steven F. Venti & David A. Wise 

- Even as people get older and wealthier, their wealth does not become more liquid. A very similar proportion of their wealth derives from housing equity, not non-housing savings. Most lower-income elderly people have low housing wealth, similar to lower-income young people. Lump-sum reverse mortgages could help increase liquidity for a small proportion of the elderly, but overall there are not many easy solutions to help increase liquid wealth for either younger people or the elderly.

Article 5:
Wealth, Composition, Housing, Income and Consumption
By: Sheng Guo & William G. Hardin III

- There is a weaker link between housing wealth and consumption compared to the link between financial wealth and consumption. People with less financial wealth were more dependent on income for their consumption. Additionally, people with more financial wealth in housing are more vulnerable to consumption changes due to economic cycles.

Article 6:
The Importance of Wealth and Income in the Transition to Homeownership
By: Zhu Xiao Di & Xiaodong Liu

- Wealth is the primary determinant if one achieves home ownership, but income also does have an effect. During the 1990s, wealth became less important to house ownership, although the importance of income stayed constant. Increases for those at lower wealth and income levels have the greatest effect. Due to the importance of wealth in house ownership, POC ownership was significantly lower.

Article 7:
Homeownership and wealth accumulation for low-income households
By: Allison Wainer & Jeffrey Zabel

- The study challenges the belief that house ownership is a pathway for lower-income people to build wealth. Increases in wealth for lower-income people due to home ownership was very dependent on timing. Purchases of homes in 2001, 2003, or 2005 had negative effects on wealth. On the other hand, there were positive effects for those who bought homes in 1994 or 1999. The increase in wealth was lower for those who purchased homes right or during a recession, while those bought a home during a time of stability saw a larger wealth increase. 

Article 8:
Financial position and house price determination: An empirical study of income and wealth effects
By: Joep Steegmans & Wolter Hassink

- This study shows that buyers and sellersâ€™ financial positions affect house prices as financially stronger sellers receive lower prices, while wealthier buyers pay more. Both income and wealth influence prices differently, with bargaining power decreasing as income rises. Wealth effects also diminish at higher wealth levels. These findings hold across market conditions and are robust to different wealth measures and bargaining assumptions.

```python
import seaborn as sns
import scipy.stats as stats
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind
import pandas as pd
from statsmodels.stats.proportion import proportions_ztest
from scipy.stats import f_oneway
```

```python
# Import data
data = pd.read_csv("/Users/kush/Data/usa_00001.csv")
```

## Data cleaning and dummy data
- Dropped rows with missing values 
- Specified AGE to include only values between 18 and 64 since those are the ages people typically work in and purhase homes at. This removes non-working age individuals who may be earning "income" from pensions, etc. 
- Excluded rows with placeholder values or non-valid entries. This included removing extreme values (< 9999998) which are used to represent missing or invalid data.

Converted the categorical variables SEX and RACE into dummy variables to improve its interpretation:

- SEX: Created Female (1 for females, 0 otherwise).
    
- RACE: Created White, Black, and Others dummies.

```python
vars = ['AGE', 'SEX', 'RACE', 'VALUEH', 'INCTOT']
pdata = data[vars]

pdata = pdata.dropna(subset=['AGE', 'SEX', 'RACE', 'VALUEH', 'INCTOT'])

pdata = pdata[(pdata['AGE'] > 18) & (pdata['AGE'] < 65)]

pdata = pdata[(pdata['INCTOT'] >= 0) & (pdata['INCTOT'] < 9999998)]
pdata = pdata[(pdata['VALUEH'] >= 0) & (pdata['VALUEH'] < 9999998)]

summarystats = pdata.describe().T[['count', 'mean', 'std', 'min', 'max']].round(4)


#Dummy data
pdata['Female'] = (pdata['SEX'] == 2).astype(int)  # 1 for Female, 0 otherwise
pdata['Male'] = (pdata['SEX'] == 1).astype(int)    # 1 for Male, 0 otherwise

pdata['White'] = (pdata['RACE'] == 1).astype(int)  # 1 for White, 0 otherwise
pdata['Black'] = (pdata['RACE'] == 2).astype(int)  # 1 for Black, 0 otherwise
pdata['Others'] = ((pdata['RACE'] != 1) & (pdata['RACE'] != 2)).astype(int)  # 1 for Other races

sexproportions = pdata[['Female', 'Male']].mean()
raceproportions = pdata[['White', 'Black', 'Others']].mean()


#results
print("Summary Statistics for Numeric Variables:")
print(summarystats)

print("\nProportions for SEX Dummy Variables:")
print(sexproportions)

print("\nProportions for RACE Dummy Variables:")
print(raceproportions)
```

```python
# Downsample to 1% of the cleaned data since the data from the actual population is too large to render the graph
sampled_data = pdata.sample(frac=0.01, random_state=1)
```

```python
# Plotting a Bar Plot that compares the average value for the INCTOT variable with the VALUEH variable
# Calculating averages for INCTOT and VALUEH
average_inctot = sampled_data['INCTOT'].mean()
average_valueh = sampled_data['VALUEH'].mean()

# Creating a summary DataFrame for plotting
summary_df = pd.DataFrame({
    'Metric': ['Total Income (INCTOT)', 'Home Value (VALUEH)'],
    'Average Value': [average_inctot, average_valueh]
})
plt.figure(figsize=(8, 6))
plt.bar(summary_df['Metric'], summary_df['Average Value'], color=['lightblue', 'lightgreen'])

# Enhancing the plot
plt.title('Comparison of Average Total Income and Home Value')
plt.ylabel('Average Value')
plt.xlabel('Metrics')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
```

```python
#Plotting a Box Plot that compares the VALUEH and the INCTOT variable on the basis of gender
# Load dataset
df_clean = sampled_data[['INCTOT', 'VALUEH', 'SEX']].dropna()

# Mapping gender codes to labels
df_clean['SEX'] = df_clean['SEX'].map({1: 'Male', 2: 'Female'})

# Reshape data for box plot
df_melted = df_clean.melt(id_vars='SEX', value_vars=['INCTOT', 'VALUEH'], var_name='Metric', value_name='Value')

# Plotting
plt.figure(figsize=(8, 6))
sns.boxplot(x='Metric', y='Value', hue='SEX', data=df_melted, palette='pastel')

# Enhancing the plot
plt.title('Distribution of Income and Home Value by Gender')
plt.ylabel('Value')
plt.xlabel('Metric')
plt.yscale('log')  # Optional: log scale for wide value ranges
plt.legend(title='Gender')
plt.tight_layout()
plt.show()
```

```python
# Hypothesis Test for difference in proportions
#Ho: There is no difference in the proportion of people with high-value homes between middle-income and upper-income groups.
#Ha: The proportion of people with high-value homes is different between the middle-income and upper-income groups.

high_income = pdata[pdata['INCTOT'] > 169800]  #PEW research for upper income
middle_income = pdata[(pdata['INCTOT'] > 56600) & (pdata['INCTOT'] <= 169800)]  #PEW research for middle income

home_value_threshold = pdata['VALUEH'].median()   # above median value is defined as high value

success_high = sum(high_income['VALUEH'] > home_value_threshold)  # High value homes in upper income group
success_mid = sum(middle_income['VALUEH'] > home_value_threshold)  # High value homes in middle income group

n_high = len(high_income)
n_mid = len(middle_income)

z_stat, p_val_prop = proportions_ztest([success_high, success_mid], [n_high, n_mid])

print(f"Hypothesis Test for difference in proportion of people with high-value homes between middle-income and upper-income groups")
print(f"Z-Statistic: {round(z_stat, 4)}")
print(f"P-Value: {round(p_val_prop, 4)}")
```

```python
# Test 2 
# Ho: The mean house value of women is equal to the mean house value of men
# Ha: The mean house value of women is not equal to the mean house value of men

high_income = pdata[pdata['INCTOT'] > 169800]  # based PEW research for upper income
middle_income = pdata[(pdata['INCTOT'] > 56600) & (pdata['INCTOT'] <= 169800)]    # based PEW research for middle income
low_income = pdata[pdata['INCTOT'] <= 56600]   # based on PEW research for lower income


# 1 for Male, 2 for Female
house_values_female = pdata[pdata['SEX'] == 2]['VALUEH']
house_values_male = pdata[pdata['SEX'] == 1]['VALUEH']

t_stat_gender, p_val_gender = ttest_ind(
    house_values_female, 
    house_values_male, 
    equal_var=False  # Assume unequal variances
)

print(f"Hypothesis Test for difference in mean house value of women vs men")
print('test-stat:', round(t_stat_gender,4))
print('p-value: ', round(p_val_gender,4))
```

```python
# Step 2: Group data by RACE
race_groups = [group['INCTOT'] for name, group in pdata.groupby('RACE')]

# Step 3: Perform ANOVA
anova_result = f_oneway(*race_groups)

# Print the ANOVA results
print("ANOVA Results by Race:")
print(f"F-statistic: {anova_result.statistic:.3f}")
print(f"P-value: {anova_result.pvalue:.3f}")

# Step 5: Interpreting the Results
if anova_result.pvalue < 0.05:
    print("There is a statistically significant difference in INCTOT across racial groups (p < 0.05).")
else:
    print("No significant difference found across racial groups.")
```

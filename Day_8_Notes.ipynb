```python
#Import packages 
import numpy as np
import pandas as pd
import seaborn as sns
import scipy.stats as stats
```

## Uniform Distribution

```python
# Generate a Vector of 5 random numbers from Unif(0,1)
np.random.seed(156)
u1 = stats.uniform.rvs(size=5)
u1
```

```python
# Generate a vector of size 30 from Unif(10,100)
# Format : stats.uniform.rvs(size,loc,scale)
np.random.seed(598)
u2 = stats.uniform.rvs(size=30, loc=10, scale=100-10)
u2
```

```python
# Probability of P(X>x): stats.unifiorm.cdf(x,a,b-a)
# Find P(X>3) on Unif(1,20)
stats.uniform.cdf(3,1,20-1)
```

```python
# Find P(X>4) on Unif(2,12)
1 - stats.uniform.cdf(4,2,12-2)
```

## Normal Distribution

```python
# Generate 30 random numbers from Norm(50, 10)
n1 = stats.norm.rvs(size = 30, loc = 50, scale = 10)
n1
```

```python
# Find P(X<20) for Norm(22,5)
stats.norm.cdf(20,22,5)
```

```python
# Find P(17.2<X<28,5) for Norm(18,6)
stats.norm.cdf(28.5,18,6) - stats.norm.cdf(17.2,18,6)
```

```python
# Find P(X>15) or P(X>25) for Norm(20,3)
stats.norm.cdf(15,20,3) + (1-stats.norm.cdf(25,20,3))
```

## Histograms and Distributions 

```python
# Generate 1,000,000 numbers from Unif(0,10)
data_unif = stats.uniform.rvs(size = 1000000, loc = 0, scale = 10)
# Generate 10,000 numbers from Norm(50,5)
data_norm = stats.norm.rvs(size = 10000, loc = 10, scale = 5)
# Generate 10,000 numbers from Binom(n=100, p= 0.5)
data_binom = stats.binom.rvs(size = 10000, n = 100, p = 0.5)
```

```python
# Plot uniform data
hist_unif = sns.histplot(data_unif);
hist_unif.set(title = 'Histogram of uniform data', xlabel='x', ylabel='frequency');
```

```python
# Plot Normal data 
hist_norm = sns.histplot(data_norm);
hist_norm.set(title = 'Histogram of normal data', xlabel='x', ylabel='frequency');
```

```python
# Plot Binomial data 
hist_binom = sns.histplot(data_binom);
hist_binom.set(title = 'Histogram of binomial data', xlabel='x', ylabel='frequency');
```

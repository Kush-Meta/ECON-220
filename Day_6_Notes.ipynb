```python
# Import necessary packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Import data
college = pd.read_csv('http://672258.youcanlearnit.net/college.csv')
```

```python
# Browse data
college.head()
```

```python
# Data's info 
college.info()
```

# Scatterplots 
- Useful to show correlation between tuition and SAT score 

```python
# plot correlation between tuition and SAT score 
sns.scatterplot(data = college, x ='tuition', y = 'sat_avg', color = 'turquoise'); # the ; is to remove text that is not necessary
```

```python
# Split the sample into two groups, public and private universities
scatter = sns.scatterplot(data = college, x ='tuition', y = 'sat_avg', hue = 'control');
# Giving title and labels
scatter.set(title = "Tuition vs SAT Score", xlabel = "Tuition", ylabel = "SAT Score")
# Change legend
plt.legend(title = 'School Type')
```

```python
# Split the sample into two plots side by side: use the "relplot()" 
scatter = sns.relplot(data = college, x ='tuition', y = 'sat_avg', hue = 'control', col = 'control');
# Giving title and labels
scatter.set(title = "Tuition vs SAT Score", xlabel = "Tuition", ylabel = "SAT Score")
# Change Legend
scatter.legend.set_title("School Type")
```

```python
# Add a trend line: use the function "lmplot"
scatter = sns.lmplot(data = college, x ='tuition', y = 'sat_avg', hue = 'control', col = 'control');
# Giving title and labels
scatter.set(title = "Tuition vs SAT Score", xlabel = "Tuition", ylabel = "SAT Score")
```

```python
# Change the color of the trend line to black
# Split the sample into two plots side by side: use the "relplot()" 
scatter = sns.lmplot(data = college, x ='tuition', y = 'sat_avg', hue = 'control', col = 'control', line_kws = {'color': 'black'});
# Giving title and labels
scatter.set(title = "Tuition vs SAT Score", xlabel = "Tuition", ylabel = "SAT Score")
```

### Bar Plots
- Compare average between groups in data

```python
# Bar plot to compare average SAT score between public and private universities
bar = sns.barplot(data = college, x = 'control', y = 'sat_avg');
bar.set(title = 'SAT score by school type', xlabel = 'School Type', ylabel = 'SAT Score')
```

```python
# Bar plot to compare average median debt among schools by highest degree offered
bar = sns.barplot(data = college, x = 'highest_degree', y = 'median_debt', errorbar = None);
# Change labels and titles
bar.set(title = 'Median debt by highest degree offered', xlabel = 'highest degree', ylabel = 'median debt')
```

```python
# Bar Plot to compare tuition by region 
# Also, split schools into public and private universities within each other
bar = sns.barplot(data = college, x = 'region', y = 'tuition', errorbar = None, hue = 'control');
# Change labels and titles
bar.set(title = 'Average tuition by region', xlabel = 'region', ylabel = 'average tuition')
```
